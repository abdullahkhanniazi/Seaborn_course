```python
# Load tip dataset
tips = sns.load_dataset('tips')
```

### Categorical Plots
#### Bar Plot

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# Customization
sns.barplot(x='day', y='total_bill', hue='sex', data=tips, palette='Blues')
plt.show()
```

```python
# Example 
# Create a random dataset using NumPy
np.random.seed(42)  # For reproducibility
Subject = ['Subject A', 'Subject B', 'Subject C', 'Subject D']
values = np.random.randint(20, 100, size=4)  # Random values between 10 and 100

# Create a DataFrame using Pandas
data = pd.DataFrame({
    'Category': Subject,
    'Value': values
})

# Display the DataFrame
print(data)

# Create the barplot using Seaborn
sns.barplot(x='Category', y='Value', data=data, palette="Blues")

# Customize the plot
plt.title("Random Bar Plot Example")
plt.xlabel("Subjects")
plt.ylabel("Values")
plt.show()
```

#### Count Plot

```python
sns.countplot(x='day', data=tips)
plt.show()
```

#### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# Example 
np.random.seed(42)
Subject = ['Subject A', 'Subject B', 'Subject C', 'Subject D', 'Subject E', 'Subject F']
values = np.random.randint(20, 50, size=6)

# Create a DataFrame using Pandas
data = pd.DataFrame({
    'Category': Subject,
    'Value': values
})
# here is good example when we have one value because this will not create any box plot box plot need 4 values like min
# max high and low
sns.boxplot(x='Category', y="Value", data=data)

# Customize the plot
plt.title("Count Plot Example")
plt.xlabel("Subjects")
plt.ylabel("Values")
plt.show()
```

### Matrix Plots
#### Heatmap

```python
df = sns.load_dataset("titanic")
data_df = df.iloc[:, [0,1,3,4,5,6]]
corelation = data_df.corr()

corelation   # Correlation basically tells you how strong the relation between 2 values it range bet. [-1,1]
plt.figure(figsize=(5,5))
sns.heatmap(corelation, cbar=True,square=True,  annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
```

### FacetGrid

```python
# Creates subplots based on the levels of a categorical variable
g = sns.FacetGrid(tips, col='sex', row='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
from sklearn.datasets import load_boston
bostan_ds = load_boston()
```

```python
houses = pd.DataFrame(bostan_ds.data, columns=bostan_ds.feature_names)
houses
```

```python
houses["PRICE"] = bostan_ds.target
```

```python
houses.head(5)

# DIstribution plot 
# sns.displot(houses["PRICE"])
# plt.show()
sns.distplot(houses["PRICE"])
plt.show()
```

```python
# Corelation method 
corelation = houses.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corelation, cbar=True,square=True,  annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
```
